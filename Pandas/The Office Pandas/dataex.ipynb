```python
import pandas as pd
import openpyxl

office = pd.read_csv('office.csv')
print(office.info())
print(office['Salary'].describe())
missing = office.isnull().sum()
print("Number of Missing values in each Column is:\n",missing)
```

```python
office['Salary'] = office['Salary'].astype(int)
sales = office[(office['Salary'] < 50000) & (office['Department'] == 'Sales')][['Name', 'Salary', 'Department']]
print(sales)
```

```python
office = office.fillna(office.mode().iloc[0])
print(office)
avg = office.groupby('Department')['Salary'].mean()
print(avg)
count_city = office['City'].value_counts()
print(count_city)
```
